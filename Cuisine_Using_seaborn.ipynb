# User Analysis

# **POV :**

It is a Countplot Representation of Gender distribution where Gender holds catagorical datas Male and Female.It is used to count Categorical Datas and Represent it into a diagrame .


Here from the polt we can understand total number of Male and Female customers without indicating any specific Cuisines in this dataset and from the diagrame or plot.

# **POV :**

It is a Histogram Plot Representation Of Overall Rating according to Cuisines where Overall Rating holds numerical datas.It is used to analyze distribution of numerical datas with the help of bins into a diagrame .


Here from the polt we can understand total number of customers and their distributions for a  specific Cuisine and follow this for other cuisines also in this dataset and from the diagrame or plot.


```python
visual = sns.histplot(x="Overall Rating", hue="Cuisines", data=dataset, bins="auto", palette="viridis")
plt.title("Histogram Plot Representation Of Overall Rating according to Cuisines", fontsize=13, loc="left")
plt.xlabel("Overall Rating")
plt.ylabel("Count of Overall Rating")
#annotate
for i in visual.patches:
  # print((i.get_x() + i.get_width()/2., i.get_height()))
  visual.annotate(
      format(i.get_height(), '.2f'),
      (i.get_x() + i.get_width()/2., i.get_height()),
      ha="center", #va="center",
      xytext=(3.5, 3.5),
      textcoords="offset points",
      fontsize=13, fontweight="normal"
  )
plt.grid(axis = "both")
plt.show()
```

# **POV:**

It is a KDE Histogram Plot Representation Of Overall Rating according to Cuisines KDE stands for Kernel Density Estimation, a method used to produce a smooth, continuous estimate of the probability density function (PDF) from a dataset. When included in a histogram plot, the KDE is typically represented as an overlaid line that visually summarizes the underlying distribution of the data, offering an alternative to the discrete bars of the histogram.

Here we can understand that kde function helps us to produce a smooth, continuous estimate of the probability density function (PDF) of the different cuisines in the dataset.  

```python
visual = sns.histplot(x="Overall Rating", hue="Cuisines", data=dataset, kde=True, bins="auto", palette="viridis")
plt.title(" KDE Histogram Plot Representation Of Overall Rating according to Cuisines", fontsize=13, loc="left")
plt.xlabel("Overall Rating")
plt.ylabel("Count of Overall Rating")
# annotate
for i in visual.patches:
  # print((i.get_x() + i.get_width()/2., i.get_height()))
  visual.annotate(
      format(i.get_height(), '.2f'),
      (i.get_x() + i.get_width()/2., i.get_height()),
      ha="center", #va="center",
      xytext=(3.5, 3.5),
      textcoords="offset points",
      fontsize=13, fontweight="normal"
  )
plt.grid(axis = "both")
plt.show()
```

## **POV :**


It is a heatmap representation of all numerical datas which are correlated. The Correlated numerical datas are User_ID, Area Code, YOB, Budget, Food Rating, Service Rating, Overall Rating. This plot representation helps us to represent Correlational data using seaborn.


**Correlation Matrix Reading**
1. If Value (0.5 to 1) then Positive Correlation
2. If Value (0.0 to 0.5) then No Correlation
3. If Value (Less Than 0.0 / -0.35454) then negetive correlation

```python
pearson_dataset_corr = dataset[["Area code","YOB","Budget","Food Rating","Service Rating","Overall Rating"]].corr(method="pearson")
```

```python
pearson_dataset_corr
```

```python
sns.heatmap(pearson_dataset_corr, cmap="viridis", annot=True)
plt.title("Heatmap Representation on Correlational Data", fontsize=13, loc="left")
plt.show()
```
